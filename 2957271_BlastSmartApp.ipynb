```python
#Question 1: Surface Blasting Design Calculator#

#Blasting Diameter (BD)
#Average In-Hole Density(p)
#Spacing To Burden Ratio (S/B)
#Spacing (S)
#Bench Height (BH)
#Explosive Mass (M)
#Sub-Drill (U)
#Stemming Length (SL)
#Burden (B)
#Technical Powder Factor (PF)

D = float(input("BD = "))
p = float(input("p = "))
a = float(input("a = "))
S = int(input("S = "))
BH = float(input("BH = "))
Rock_Type = str(input("Rock Type (Hard, Medium, Soft, Very Soft):"))

U = 1/4*D
print("U = ", U)
M = (p*(D**2))/1273
print("M = ", M)
```

```python
#Stemming Length Calculation
SL = round(20*D, 2)
print("SL = ", SL)
#Technical Powder Factor
if Rock_Type == "Hard":
    Tech_PF = 0.7
elif Rock_Type == "Medium":
    Tech_PF = 0.45
elif Rock_Type == "Soft":
    Tech_PF = 0.3
elif Rock_Type == "Very Soft":
    Tech_PF = 0.15
else:
    Tech_PF = None
    print("Invalid Rock Type")
if Tech_PF is not None:
    print("Technical Powder Factor =", Tech_PF)

while True:
    if SL >= BH:
        print("Error! SL cannot be greater than BH")
        BH = float(input("BH = "))
    else:
        break
```

```python
#Burden Calculation
import math
B = float(round(math.sqrt((M*(BH-SL))/(a*Tech_PF*BH)), 2))
print("Burden = ", B)

#Powder Factor Calculation
Actual_PF = float(round((BH-SL+U)*M/(B*S*BH), 4))
print("Actual Powder Factor = ", Actual_PF)
```
